```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
wine = load_wine()
```

```python
data = pd.DataFrame(wine.data, columns=wine.feature_names)
data['label'] = wine.target
```

```python
data.head()
```

```python
data['label'].value_counts()
```

```python
wine.feature_names
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(data.corr(), annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
plt.title('Feature Correlation Heatmap')
plt.show()

```

```python
X = wine.data
y = wine.target
```

```python
splits = ['85-15', '70-30', '60-40']
wine_acc = []
iris_acc = []
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, shuffle=True)#80-20 split
```

```python
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
acc1=accuracy_score(y_test, y_pred)
print("Accuracy:", acc1)
print(classification_report(y_test, y_pred))
wine_acc.append(acc1)
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 3))
sns.heatmap(cm, annot=True,cmap='YlGnBu')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target # Add target labels
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2:'virginica'})
```

```python
df.head(6)
```

```python
df['species'].value_counts()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'DejaVu Sans'  # or simply remove this line

plt.figure(figsize=(10, 6))
sns.heatmap(df.iloc[:, :-1].corr(), annot=True, cmap='YlGnBu')
plt.title("Iris Feature Correlation")
plt.show()
```

```python
X=iris.data
y=iris.target
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,shuffle=True)#80-20 split
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
acc1=accuracy_score(y_test, y_pred)
print("Accuracy:", acc1)
print(classification_report(y_test, y_pred))
iris_acc.append(acc1)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 3))
sns.heatmap(cm, annot=True,cmap='YlGnBu')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title('Confusion Matrix')
plt.show()
print("\nFirst 8 predictions:", model.predict(x_test[:8]))

custom_input = np.array([6.9, 3.2, 5.7, 2.3]).reshape(1, -1)
custom_output = model.predict(custom_input)
print("\nCustom sample prediction:", custom_output[0], f"({iris.target_names[custom_output[0]]})")
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.25,shuffle=True,stratify=y)#75-25 split
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
acc2=accuracy_score(y_test, y_pred)
print("Accuracy:", acc2)
print(classification_report(y_test, y_pred))
iris_acc.append(acc2)
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 3))
sns.heatmap(cm, annot=True,cmap='YlGnBu')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title('Confusion Matrix')
plt.show()
print("\nFirst 8 predictions:", model.predict(x_test[:8]))

custom_input = np.array([6.9, 3.2, 5.7, 2.3]).reshape(1, -1)
custom_output = model.predict(custom_input)
print("\nCustom sample prediction:", custom_output[0], f"({iris.target_names[custom_output[0]]})")
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3,shuffle=True,stratify=y)#70-30 split
model = GaussianNB()
model.fit(x_train, y_train)
```
