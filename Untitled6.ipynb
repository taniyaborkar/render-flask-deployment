```python
from sklearn.datasets import load_wine
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
wine=load_wine()
```

```python
data= pd.DataFrame(wine.data,columns=wine.feature_names)
data.head()
```

```python
data= pd.DataFrame(wine.data,columns=wine.feature_names)
data.head()
```

```python
x=wine.data
y=wine.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.25,shuffle=True)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy',max_depth=5,min_samples_leaf=5,random_state=40)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
print("Accuracy_we:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(clf, feature_names=wine.feature_names, class_names=wine.target_names, filled=True)
plt.show()
```

```python
feature_importances=pd.Series(clf.feature_importances_,index=wine.feature_names)
print('Feature importance:')
print(feature_importances.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
data = pd.DataFrame(iris.data, columns=iris.feature_names)
print(data.head())
```

```python
x = iris.data
y = iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3,shuffle=True)
```

```python

clf1 = tree.DecisionTreeClassifier(criterion='entropy', max_depth=4, min_samples_leaf=5,random_state=45)
```

```python
clf1.fit(X_train,y_train)
```

```python
y_pred=clf1.predict(X_test)
```

```python

from sklearn import tree
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report

# Assuming y_test and y_pred are already defined and clf1 is fitted
print("Accuracy_ig:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Plotting the decision tree with smaller font sizes
plt.figure(figsize=(10, 8))
tree.plot_tree(clf1,
               feature_names=iris.feature_names,
               class_names=iris.target_names,
               filled=True,
               fontsize=10)  # Adjust fontsize here

plt.show()
```

```python
feature_importances = pd.Series(clf1.feature_importances_, index=iris.feature_names)
print('Feature importance:')
print(feature_importances.sort_values(ascending=False))
```

```python
import matplotlib.pyplot as plt
import pandas as pd

wine_accuracies = {
    'entropy': 0.9359259259259259,
    'log_loss': 0.9174074074074074,
    'gini': 0.9718518518518519
}
iris_accuracies = {
    'entropy': 0.9177777777777777,
    'log_loss': 0.9533333333333333,
    'gini': 0.9755555555555556
}

accuracy_df = pd.DataFrame({
    'Criterion': ['entropy', 'log_loss', 'gini'],
    'Wine Dataset': list(wine_accuracies.values()),
    'Iris Dataset': list(iris_accuracies.values())
})

fig, ax = plt.subplots(figsize=(10, 6))
accuracy_df.set_index('Criterion', inplace=True)

# Plotting the data as a line graph
accuracy_df.plot(kind='line', ax=ax, colormap='viridis', marker='o', linestyle='-', linewidth=2, markersize=8)

# Annotating the data points
for line in ax.lines:
    for x, y in zip(line.get_xdata(), line.get_ydata()):
        ax.annotate(f'{y*100:.2f}%', (x, y), textcoords="offset points", xytext=(0, 5), ha='center', fontsize=10)

# Set y-axis limits and ticks
ax.set_ylim(0.80, 1.00)  # Set the y-axis range from 80% to 100%
ax.set_yticks([i/100 for i in range(80, 101, 5)])  # Set y-ticks in increments of 5%

plt.title('Accuracy Comparison for Different Criteria on Wine and Iris Datasets')
plt.ylabel('Accuracy')
plt.xlabel('Criterion')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
