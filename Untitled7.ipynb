```python
import csv
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
sales_data = [
    ['Visits', 'Purchase Amount'],
    [10, 20], [9, 0], [6, 5], [7, 25], [11, 0],
    [12, 10], [8, 11], [4, 150], [5, 120], [7, 100],
    [5, 200], [6, 120], [4, 220], [6, 300], [7, 404],
    [5, 388], [6, 225], [8, 350], [8, 236], [4, 167],
    [9, 400], [9, 699], [5, 553], [8, 450], [9, 817],
    [1, 1010], [11, 825], [11, 700]
]
```

```python
with open('sales.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(sales_data)
```

```python
df_sales = pd.read_csv('sales.csv')
X_sales = df_sales[['Visits', 'Purchase Amount']].values
silhouette_scores = []
k_values = [6, 7, 8]
print("Silhouette (Accuracy) Scores:\n")
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=0, n_init=10)
    labels = kmeans.fit_predict(X_sales)
    score = silhouette_score(X_sales, labels)
    silhouette_scores.append(score)
    print(f"Accuracy (Silhouette Score) for k={k}: {score:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.plot([str(k) for k in k_values], silhouette_scores, marker='o', color='Green', linestyle='-', linewidth=2)
plt.title('Accuracy Comparison (Silhouette Scores) for k=6,7,8')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score (Accuracy)')
plt.grid(True, axis='y')
plt.show()
```

```python
k = 6
kmeans_sales = KMeans(n_clusters=k, random_state=0, n_init=4)

# Fit and predict labels
labels_sales = kmeans_sales.fit_predict(X_sales)

# Get centroids and convert to integers
centroids_sales = np.round(kmeans_sales.cluster_centers_).astype(int)

print(f"\nCluster Centers for k={k} (as integers):\n{centroids_sales}")
print(f"Cluster Labels:\n{labels_sales}")
```

```python
colors_sales = ['black', 'orange', 'pink', 'green', 'blue', 'red']
markers_sales = ['o', 's', '^', 'D', 'v', 'P']


fig, ax = plt.subplots(figsize=(8, 4))
for i in range(k):
    points = X_sales[labels_sales == i]
    ax.scatter(points[:, 0], points[:, 1], s=50, c=colors_sales[i % len(colors_sales)],
           marker=markers_sales[i % len(markers_sales)], label=f'Cluster {i}')

    ax.scatter(centroids_sales[i, 0], centroids_sales[i, 1], marker='*', s=300,
               c=colors_sales[i], edgecolors='black')
xnew_sales = np.array([[4, 210], [7, 500], [10, 150], [2, 18]])

newlabels_sales = kmeans_sales.predict(xnew_sales)
ax.scatter(xnew_sales[:, 0], xnew_sales[:, 1], s=250, marker='X',
           c=[colors_sales[i] for i in newlabels_sales], edgecolors='black')

ax.set_xlabel('Number of Visits')
ax.set_ylabel('Purchase Amount')
ax.set_title('Sales Data - KMeans Clustering (k=4)')
ax.grid(True)
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
