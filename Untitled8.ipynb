```python
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVC
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score
```

```python
data = load_wine()
X, y = data.data, data.target
```

```python
svm_configs = {
    "SVM (Linear, C=1)": make_pipeline(StandardScaler(), SVC(kernel="linear", C=1)),
    "SVM (Linear, C=10)": make_pipeline(StandardScaler(), SVC(kernel="linear", C=10)),
    "SVM (RBF, C=1, gamma='scale')": make_pipeline(StandardScaler(), SVC(kernel="rbf", C=1, gamma="scale")),
    "SVM (RBF, C=10, gamma=0.1)": make_pipeline(StandardScaler(), SVC(kernel="rbf", C=10, gamma=0.2))
}
```

```python
svm_simple = make_pipeline(StandardScaler(), SVC(kernel="linear", C=1))
svm_simple.fit(X, y)
y_pred_simple = svm_simple.predict(X)
acc_simple = accuracy_score(y, y_pred_simple)
print(f"Simple SVM Test Accuracy (Without K-Fold): {acc_simple:.4f}")
```

```python
k = 4 # 4-Fold Cross Validation
kf = KFold(n_splits=k, shuffle=True, random_state=45)
best_model = None
best_score = 0
print("\nSVM Model Performance (Mean Accuracy across 4 folds):")
for name, model in svm_configs.items():
    scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
    mean_score = np.mean(scores)
    print(f"{name}: {mean_score:.4f}")
    if mean_score > best_score:
        best_score = mean_score
        best_model = model
print("\nBest SVM Model Selected:", best_model.named_steps['svc'])
```

```python
best_model.fit(X, y)
```

```python
new_sample = X[1].reshape(1, -1)  # Using the first sample as a test
prediction = best_model.predict(new_sample)
print("\nPredicted class for new sample:", data.target_names[prediction[0]])
```

```python
k = 5  # 5-Fold Cross Validation
kf = KFold(n_splits=k, shuffle=True, random_state=42)
best_model = None
best_score = 0
print("\nSVM Model Performance (Mean Accuracy across 5 folds):")
for name, model in svm_configs.items():
    scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
    mean_score = np.mean(scores)
    print(f"{name}: {mean_score:.4f}")
    if mean_score > best_score:
        best_score = mean_score
        best_model = model
print("\nBest SVM Model Selected:", best_model.named_steps['svc'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Accuracies
iris_scores = [0.9667,  0.9664, 0.9733, 0.9667]
wine_scores = [1.0000,  0.9775, 0.9830, 0.9775]
labels = ['Simple SVM', 'K=4', 'K=5', 'K=6']

x = np.arange(len(labels))  # numerical x positions

fig, ax = plt.subplots(figsize=(10, 6))

# Plot lines
ax.plot(x, iris_scores, marker='o', linestyle='-', label='Iris', color='skyblue')
ax.plot(x, wine_scores, marker='s', linestyle='-', label='Wine', color='lightgreen')

# Annotate points
for i, score in enumerate(iris_scores):
    ax.annotate(f'{score:.4f}', (x[i], iris_scores[i]), textcoords="offset points", xytext=(0, 6), ha='center', fontsize=9)

for i, score in enumerate(wine_scores):
    ax.annotate(f'{score:.4f}', (x[i], wine_scores[i]), textcoords="offset points", xytext=(0, 6), ha='center', fontsize=9)

ax.set_ylabel('Accuracy')
ax.set_title('SVM Accuracy Comparison (Simple vs K-Folds)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.set_ylim(0.94, 1.02)
ax.legend()
ax.grid(axis='y', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
